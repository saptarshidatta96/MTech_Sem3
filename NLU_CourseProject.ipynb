```python
json_file = "/content/gdrive/MyDrive/train.txt"
```

```python
import requests

import json

with open(json_file, "r") as f:
    content = f.read()
parsed_values = []
decoder = json.JSONDecoder()
while content:
    value, new_start = decoder.raw_decode(content)
    content = content[new_start:].strip()
    # You can handle the value directly in this loop:
    # Or you can store it in a container and use it later:
    parsed_values.append(value)
```

```python
chatbot_data = []
for item in parsed_values:
  annotated_data = item['passages'][0]['exhaustivelyAnnotatedProperties']
  for i in range(len(annotated_data)):
    if annotated_data[i]['propertyName'] == 'DATE_FOUNDED':
      facts = item['passages'][0]['facts']
      for relation in range(len(facts)):
        if facts[0]['humanReadable'].split('>')[1] == ' <DATE_FOUNDED':
          chatbot_data.append((facts[0]['humanReadable'].split('>')[0].replace('<', ''), facts[0]['humanReadable'].split('>')[1].replace('<', ''), 
                              facts[0]['humanReadable'].split('>')[2].replace('<', '')))
          
          
    if annotated_data[i]['propertyName'] == 'HEADQUARTERS':
      facts = item['passages'][0]['facts']
      for relation in range(len(facts)):
        if facts[0]['humanReadable'].split('>')[1] == ' <HEADQUARTERS':
          chatbot_data.append((facts[0]['humanReadable'].split('>')[0].replace('<', ''), facts[0]['humanReadable'].split('>')[1].replace('<', ''), 
                              facts[0]['humanReadable'].split('>')[2].replace('<', '')))
          

    if annotated_data[i]['propertyName'] == 'SUBSIDIARY_OF':
      facts = item['passages'][0]['facts']
      for relation in range(len(facts)):
        if facts[0]['humanReadable'].split('>')[1] == ' <SUBSIDIARY_OF':
          chatbot_data.append((facts[0]['humanReadable'].split('>')[0].replace('<', ''), facts[0]['humanReadable'].split('>')[1].replace('<', ''), 
                              facts[0]['humanReadable'].split('>')[2].replace('<', '')))
          

    if annotated_data[i]['propertyName'] == 'FOUNDED_BY':
      facts = item['passages'][0]['facts']
      for relation in range(len(facts)):
        if facts[0]['humanReadable'].split('>')[1] == ' <FOUNDED_BY':
          chatbot_data.append((facts[0]['humanReadable'].split('>')[0].replace('<', ''), facts[0]['humanReadable'].split('>')[1].replace('<', ''), 
                              facts[0]['humanReadable'].split('>')[2].replace('<', '')))
          

    if annotated_data[i]['propertyName'] == 'CEO':
      facts = item['passages'][0]['facts']
      for relation in range(len(facts)):
        if facts[0]['humanReadable'].split('>')[1] == ' <CEO':
          chatbot_data.append((facts[0]['humanReadable'].split('>')[0].replace('<', ''), facts[0]['humanReadable'].split('>')[1].replace('<', ''), 
                              facts[0]['humanReadable'].split('>')[2].replace('<', '')))
          

```
